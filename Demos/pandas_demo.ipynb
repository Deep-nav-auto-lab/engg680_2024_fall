```python
# Create a Python list that holds the names of the four columns.
my_column_names = ['Eleanor', 'Chidi', 'Tahani', 'Jason']

# Create a 3x4 numpy array, each cell populated with a random integer.
my_data = np.random.randint(low=0, high=101, size=(10, 4))

# Create a DataFrame.
df = pd.DataFrame(data=my_data, columns=my_column_names)

# Create a column named Janet whose contents are the sum
# of two other columns.
df['Janet'] = df['Tahani'] + df['Jason']

# Print the enhanced DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()

df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["A", "B", "C", "D"])
df = df.cumsum()
plt.figure()
df.plot()
plt.legend(loc='best')
#
```

```python
import matplotlib.pyplot as plt

ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))

df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["A", "B", "C", "D"])

plt.figure()
df.plot()
plt.legend(loc='best')
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with random integers between 0 and 5, dimensions 10x7
# Assigning column names for better understanding and data manipulation
column_names = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
df = pd.DataFrame(np.random.randint(0, 5, (10, 7)), columns=column_names)

# Save the DataFrame to a CSV file
# Including index=False to avoid writing row indices in the file
df.to_csv("foo.csv", index=False)

# Read the CSV file back into a DataFrame
# This read operation will use the column names provided in the CSV file
read_df = pd.read_csv("foo.csv")

# Print the DataFrame that was read from the CSV file
print(read_df)
```
