```python
# Creating a dictionary
person = {
    'name': 'John',
    'age': 30,
    'city': 'New York'
}
print(person)
```

```python
# Accessing elements
print(person['name'])  # Output: John
print(person.get('age'))  # Output: 30
# Using get with default value
print(person.get('profession', 'Unknown'))  # Output: Unknown
```

```python
# Adding a new key-value pair
person['profession'] = 'Engineer'

# Changing an existing value
person['age'] = 35
```

```python
# Removing an element using del
del person['city']

# Removing an element using pop
age = person.pop('age')  # Removes 'age' and returns the value (35)

# Removing the last inserted item using popitem (Python 3.7+)
item = person.popitem()  # Removes and returns the last key-value pair

print(person)
```

```python
# Loop through the keys
for key in person:
    print(key, person[key])
# Loop through keys and values
for key, value in person.items():
    print(key, value)
```

```python
# Create a dictionary with keys as numbers and values as their squares
squares = {x: x*x for x in range(6)}
print(squares)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
# Nested dictionaries
family = {
    'person1': {'name': 'John', 'age': 30},
    'person2': {'name': 'Jane', 'age': 25},
    'person3': {'name': 'Sally', 'age': 22}
}
print(family)
```
