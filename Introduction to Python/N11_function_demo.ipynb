```python
# Define a function to add two numbers
def add(x, y):
    """Return the addition of x and y."""
    return x + y

# Define a function to subtract two numbers
def subtract(x, y):
    """Return the subtraction of y from x."""
    return x - y

# Define a function to multiply two numbers
def multiply(x, y):
    """Return the multiplication of x and y."""
    return x * y

# Define a function to divide one number by another
def divide(x, y):
    """Return the division of x by y. Handles division by zero."""
    if y == 0:
        return "Error: Division by zero."
    else:
        return x / y

# Main program to use the arithmetic functions
def main():
    print("Addition of 5 and 3:", add(5, 3))
    print("Subtraction of 5 and 3:", subtract(5, 3))
    print("Multiplication of 5 and 3:", multiply(5, 3))
    print("Division of 5 by 3:", divide(5, 3))

# Execute the main function if this file is run as a script
if __name__ == "__main__":
    main()
```

```python
import math

def area_of_circle(radius):
    """Calculate and return the area of a circle given its radius."""
    return math.pi * radius ** 2
```

```python
def calculate_compound_interest(principal, annual_rate, compounds_per_year=4, years=5, **kwargs):
    """
    Calculate the future value of an investment based on the compound interest formula.

    Parameters:
    - principal (float): The initial amount of money invested.
    - annual_rate (float): The annual interest rate as a decimal (e.g., 0.05 for 5%).
    - compounds_per_year (int, optional): How often the interest is compounded per year. Default is 4.
    - years (int, optional): The number of years the investment is to be held. Default is 5.
    - kwargs (dict): Additional keyword arguments that are not necessary for calculation but can be used for additional features like logging.

    Returns:
    - float: The future value of the investment.
    """
    # Handle any additional keyword arguments
    if 'log' in kwargs:
        print(f"Logging enabled: {kwargs['log']}")

    # Compound interest formula
    future_value = principal * (1 + annual_rate / compounds_per_year) ** (compounds_per_year * years)
    return future_value

# Example usage with default arguments
print("Future value with default arguments:", calculate_compound_interest(1000, 0.05))

# Example usage with keyword arguments
print("Future value using keyword arguments:",
      calculate_compound_interest(principal=2000, annual_rate=0.03, years=10, compounds_per_year=12))

# Example using arbitrary kwargs for additional functionality
test_dict =  {'log': "Transaction Completed with different method"}

print("Future value with logging:",
      calculate_compound_interest(1500, 0.04, log ="Transaction Completed"))

print("Future value with logging:",
      calculate_compound_interest(1500, 0.04, 4, 4, **test_dict))
```

```python
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8


numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Outputs: [1, 4, 9, 16]


numbers = [1, 2, 3, 4]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)  # Outputs: [2, 4]
```

```python
# Define a lambda function for addition. This lambda takes two parameters, x and y, and returns their sum.
add = lambda x, y: x + y
# Call the lambda function with arguments 5 and 3, then print the result. This demonstrates basic arithmetic with lambda.
print(add(5, 3))  # Output: 8

# Preparing a list of numbers to use with map and filter functions.
numbers = [1, 2, 3, 4]

# Using 'map' with a lambda function to square each number in the list 'numbers'.
# The lambda function takes an argument x and returns x squared (x^2).
squared = list(map(lambda x: x**2, numbers))
# Print the list of squared numbers. Demonstrates transforming each item in a collection.
print(squared)  # Outputs: [1, 4, 9, 16]

# Using 'filter' with a lambda function to extract even numbers from the list 'numbers'.
# The lambda function takes an argument x and checks if x is divisible by 2 (even number check).
even = list(filter(lambda x: x % 2 == 0, numbers))
# Print the list of even numbers. This filters the collection based on a condition.
print(even)  # Outputs: [2, 4]
```

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator
counter = count_up_to(5)
for num in counter:
    print(num)  # Outputs 1, 2, 3, 4, 5 sequentially
```

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with arguments {args} and keyword arguments {kwargs}")
        return func(*args, **kwargs)
    return wrapper

@logger
def add(x, y):
    return x + y

print(add(5, 3))
```
